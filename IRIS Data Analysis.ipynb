```python
# TASK 2- Find the max or min  sepal length by analysis and viz?
```

```python
max_sepal_len=df.loc[df.sepal_length.max()==df.sepal_length][['species','sepal_length']]
max_sepal_len
```

```python
min_sepal_len=df.loc[df.sepal_length.min()==df.sepal_length][['species','sepal_length']]
min_sepal_len
```

```python
# max len of sepal by viz.
```

```python
sns.jointplot(x='sepal_length',y='species',data=df)
```

```python
#result----According to thr graph max len of sepal is"VIRGINICA"  and min len of sepal is"SETOSA"
```

```python
# TASK 3 Find the max or min  sepal width by analysis and viz?
```

```python
max_sepal_width=df[df['sepal_width'].max()==df['sepal_width']][['sepal_width','species']]
max_sepal_width
```

```python
# max or min width of sepal by viz
```

```python
sns.jointplot(x='sepal_width',y='species',data=df)
```

```python
# TASK 4-Find max or min petal length  by analysis and viz?

max_petal_len=df.loc[df.petal_length.max()==df.petal_length][['petal_width','species']]
max_petal_len
```

```python
min_petal_len=df.loc[df.petal_length.min()==df.petal_length][['petal_width','species']]
min_petal_len
```

```python
# TASK 5  Find max or min petal width by analysis and viz.

max_petal_width=df[df.petal_width.max()==df.petal_width][['petal_width','species']].head(1)
max_petal_width
```

```python
min_petal_width=df[df.petal_width.min()==df.petal_width][['petal_width','species']].head(1)
min_petal_width
```

```python
# TASK 6 show the viz total number of sepal length and width petal length and width of SETOSA?
```

```python
x1=df[df['species']=='setosa']['sepal_length'].sum()
x2=df[df['species']=='setosa']['sepal_width'].sum()
x3=df[df['species']=='setosa']['petal_length'].sum()
x4=df[df['species']=='setosa']['petal_width'].sum()
data=[x1,x2,x3,x4]
mylabels=['sepal_len','sepal_width','petal_len','petal_width']
plt.pie(data,labels=mylabels,autopct='%.2f')
```

```python
#result-----
```
