```python
# Matplotlib----mathnatical plotting

import matplotlib.pyplot as plt
import numpy as np
```

```python
#line
#bar
#histogram
#pie
#subplot


#line graph---------

sales=np.array([23,56,43,89,20])
month=np.array([3,4,5,6,7])
plt.xlabel('month')
plt.ylabel('sales')
plt.plot(month,sales)
```

```python
# wap to make a temp using list and try to viz. the using data matplotlob

temp=[30,45,20,57,34,40]
days=['MON','TUE','WED','THRU','FRI','SAT']
plt.xlabel('DAYS')
plt.ylabel('TEMP')
plt.plot(days,temp)
```
