```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# line 
# bar
# histogram 
#pie
# subplot

sales=np.array([23,100,78,90])
months=np.array([4,5,6,7])
plt.xlabel('MONTHS')
plt.ylabel('SALES')
plt.plot(months,sales)
```

```python
# WAP TO MAKE A TEMP USING LIST AND TRY TO VIZ THE DATA USING MATPLOTLIB

temp=[45,20,46,90,76]
days=['sun','mon','tue','wed','thru']
plt.xlabel('DAYS')
plt.ylabel('TEMP')
plt.plot(days,temp)
```

```python
# barplot-----
sub=['ML','AI','SQL','POWER BI']
C1=[23,69,40,70]
plt.xlabel('SUBJECT')
plt.ylabel('NO_OF_STUDENT')
plt.bar(sub,C1)
```

```python
sub=['ML','AI','SQL','POWER BI']
C2=[10,50,80,20]
plt.xlabel('SUBJECT')
plt.ylabel('NO_OF_STUDENT')
plt.bar(sub,C2,color=['r','g','b','y'])
```

```python
# horizontal bar------
sub=['ML','AI','SQL','POWER BI']
C3=[70,50,70,30]
plt.xlabel('SUBJECT')
plt.ylabel('NO_OF_STUDENT')
plt.barh(sub,C3,color=['b','g','r','y'])
```

```python
sub=['ML','AI','SQL','POWER BI']
C1=[23,69,40,70]
plt.xlabel('SUBJECT')
plt.ylabel('NO_OF_STUDENT')
plt.bar(sub,C1,color='r',align='edge')
```
